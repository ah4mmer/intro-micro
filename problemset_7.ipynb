## Precursors

```python
# Def vars
var('a,h')

# Def prices and target output
p = 10      # price for a
w = 5       # price for h
qbar = 64 

# Def production function
q(a,h) = 2*a^(0.5)*h

# Def isocost line
c(a,h) = p*a + w*h
```

## Calculate equilibrium

```python
# Get MRS
MRS(a,h) = q.diff(a)/q.diff(h)
print(MRS(a,h).full_simplify())

# Equilibrium condition (-MRS = -p/w)
H(a) = solve(MRS(a,h) == p/w, h)
print('Optimum condition :: h =', H[0](a).rhs())

# Plug into production function to get a* and h*
solve_a = solve(qbar==q(a, H[0](a).rhs()), a)
opt_a = solve_a[2].rhs()

solve_h = H(opt_a)
opt_h = solve_h[0].rhs()

print('a* =', opt_a)
print('h* =', opt_h)
```

```python
# Plot equilibrium condition
equilibrium = H[0](a).rhs()
G1 = plot(equilibrium, (a,0,30), ymin=0, ymax=50, color='green', thickness=0.5, axes_labels=['$a$', '$h$'])
G1
```

```python
# Quantity and cost in equilibrium
Y = q(opt_a, opt_h)
TC = p*opt_a + w*opt_h
print('Y =', Y, '; Cost =', TC)
```

## Plot everything

```python
isocostcurve = solve(TC == c(a,h), h)[0].rhs()
print(isocostcurve)

isocostcurve_null = solve(isocostcurve == 0, a)[0].rhs()

isoquante = solve([Y == q(a,h)], h)[0].rhs()
print(isoquante)

G2 = plot(isocostcurve, (a,0,isocostcurve_null), ymin=0, ymax=50, thickness=2)
G3 = plot(isoquante, (a,0,30), ymin=0, ymax=50, color='red', thickness=2)
G4 = line([(0,opt_h), (opt_a,opt_h)], linestyle=":", color='black', thickness=0.8)
G5 = line([(opt_a,0), (opt_a,opt_h)], linestyle=":", color='black', thickness=0.8)
G6 = text("$(a^{*},h^{*})$", (opt_a+2.5, opt_h+1), fontsize=11, color='black')
Gfull = G1 + G2 + G3 + G4 + G5 + G6
Gfull
```
